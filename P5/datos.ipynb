```python
import pandas as pd


def intervalos(arr, n):
  maximo = max(arr)
  minimo = min(arr)
  rango = maximo - minimo
  intervalo = rango / n
  return tuple((round(minimo + i * intervalo, 3), round(minimo + (i+1) * intervalo, 3)) for i in range(n))


def reg_discretizado(data, label):
  df = pd.DataFrame({label: intervalos(data, 5), 'loan_status_0': 0, 'loan_status_1': 0})
  max_data = data.max()
  for i, (li, ls) in enumerate(df[label]):
    loan_status = pd.get_dummies(dataset[[label, 'loan_status']], columns=['loan_status']).query(f'{li} <= {label} {"<=" if ls == max_data else "<"} {ls}')[['loan_status_0', 'loan_status_1']].sum()
    df.iloc[i, 1] = loan_status.iloc[0]
    df.iloc[i, 2] = loan_status.iloc[1]
  return df

dataset = pd.read_csv('./credit_risk_dataset.csv').fillna(0)
```
