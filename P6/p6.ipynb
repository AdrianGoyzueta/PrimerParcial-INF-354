```python
import matplotlib.pyplot as plt


def graficar(data, intervalo, label):
  x = range(len(data["label"]))
  fig, ax = plt.subplots()
  loan_status_0 = ax.bar(x, data["0"], intervalo, label='loan_status 0')
  loan_status_1 = ax.bar([p + intervalo for p in x], data["1"], intervalo, label='loan_status 1')

  ax.set_xlabel(label)
  ax.set_ylabel('Cantidad')
  ax.set_title(f'Regla Decicion {label}')
  ax.set_xticks([p + intervalo / 2 for p in x])
  ax.set_xticklabels(data["label"], rotation=15)
  ax.legend()

  # plt.savefig(f'Regla_Decicion_{label}.png')  # Guardar imagen
  plt.show()
```
